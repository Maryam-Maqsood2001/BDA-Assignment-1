```python
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/My Drive/household_power_consumption.xlsx'
df = pd.read_excel(file_path)


```

```python
df['DateTime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'].astype(str))

df.set_index('DateTime', inplace=True)

start_date = '2007-02-01'
end_date = '2007-02-02'
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]


plt.hist(filtered_df['Global_active_power'], bins=20, color='blue', edgecolor='black')
plt.title('Histogram of Global Active Power')
plt.xlabel('Global Active Power (kW)')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
plt.savefig('plot1.png', dpi=100)
plt.close()
```
